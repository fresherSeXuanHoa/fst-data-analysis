```python
import pandas as pd
import firebase_admin
from firebase_admin import credentials, firestore

collection_name = 'tbl-19520101'
dataset_path = '../conf/database/orginal_sales.csv'
fire_store_auth_json = '../conf/auth/iuh-19520101-2f8f0-firebase-adminsdk-if2dg-355b4c424e.json'

sales_df = pd.read_csv(dataset_path)
credential = credentials.Certificate(fire_store_auth_json)
app = firebase_admin.initialize_app(credential)
database = firestore.client()
```

```python
for sale in sales_df.to_dict(orient='records'):
    document = database.collection(collection_name).document()
    document.set(sale)
```

```python
sales_firestone_df = pd.DataFrame(list(map((lambda sale : sale.to_dict()), database.collection(collection_name).stream())))
print(f'Dataset Shape: {sales_firestone_df.shape}')
```
